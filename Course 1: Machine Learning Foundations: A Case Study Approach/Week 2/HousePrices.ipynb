```python
import turicreate
```

```python
sf=turicreate.SFrame('Lecture/home_data.sframe')
```

```python
sf
```

```python
turicreate.show(sf[:5000]['sqft_living'],sf[:5000]['price'])
```

```python
 train,test=sf.random_split(0.8,seed=0)
```

```python
model=turicreate.linear_regression.create(train,target='price',features=['sqft_living'])
```

```python
model.evaluate(test) #max error on the test dataset and the root mean square error #very high maxerror in comparison to training set
```

```python
model.evaluate(train)
```
