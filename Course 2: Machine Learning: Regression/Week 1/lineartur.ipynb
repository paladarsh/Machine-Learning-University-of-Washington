```python
import turicreate
```

```python
import matplotlib.pyplot as mtl
```

```python
data=turicreate.SFrame("Crime.csv")
```

```python
mtl.plot(data['CrimeRate'],data['HousePrice'],'.')
```

```python
model=turicreate.linear_regression.create(data,target='HousePrice',features=['CrimeRate'],validation_set=None)
```

# Plotting the predicted linear regression line

```python
mtl.plot(data['CrimeRate'],data['HousePrice'],'.',data['CrimeRate'],model.predict(data),'-')
```

 # Removing the Center City points

```python
data=data[data['MilesPhila']!=0.0]
```

```python
modelwcc=turicreate.linear_regression.create(data,target="HousePrice",features=['CrimeRate'],validation_set=None)
```

```python
mtl.plot(data['CrimeRate'],data['HousePrice'],'.',data['CrimeRate'],modelwcc.predict(data),'-')
```

```python
mtl.plot(data['CrimeRate'],data['HousePrice'],'.',data['CrimeRate'],model.predict(data),'-',data['CrimeRate'],modelwcc.predict(data),'-')
```
